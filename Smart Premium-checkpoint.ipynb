```python
#Hypothesis testing

#differentiate contineous and catogory column
continues = ['Age','Health Score','Credit Score','Annual Income','Premium Amount']
categories = ['Gender', 'Marital Status','Number of Dependents','Education Level','Occupation','Location','Policy Type','Previous Claims','Vehicle Age','Insurance Duration','Customer Feedback','Smoking Status','Exercise Frequency','Property Type','Policy_Start_Date_Clean']
```

```python
from scipy import stats
```

```python
#continious vs continious
def two_sample(d1,d2):
    t=0
    f=0
    for i in range(31):
        sample1 = d1.sample(frac=0.03)
        sample2 = d2.sample(frac=0.03)
        t_test,p_value = stats.ttest_ind(sample1,sample2)
        if p_value < 0.05:
            f+=1
        else:
            t+=1
    if t>f:
        return True
    else:
        return False
    
#cat vs cat
def chisquare_test(d1,d2):
    return True if stats.chi2_contingency(pd.crosstab(d1,d2))[1]<0.05 else False

#continious(d1) vs cat(d2)
def annova_test(d1,d2):
    group = df[d2].unique()
    data={}
    for i in group:
        data[i]= df[d1][df[d2]==i]
    fvalue,p_value = stats.f_oneway(*[i for i in data.values()])
    return True if p_value>0.05 else False
```

```python
final = {}
for i in df.columns:
    final[i] = {}
    for j in df.columns:
        if (i in continues) and (j in continues):
            result = two_sample(df[i],df[j])
        elif (i in categories) and (j in categories):
            result = chisquare_test(df[i],df[j])
        elif (i in continues) and (j in categories):
            result = annova_test(i,j)
        elif (i in categories) and (j in continues):
            result = annova_test(j,i)
        if result:
            final[i][j] = 1
        else:
            final[i][j] = 0
```

```python
#Build and Train the Pipeline

#1. Import the Required Libraries
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# Models
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor

#2. Train–Test Split (Ensure True Separation)
from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test = train_test_split(df_scaled_x,y,test_size=0.2)

#3. Build Pipelines for Each Model
pipelines = {
    'Linear Regression': Pipeline([
        ('model', LinearRegression())
    ]),
    
    'Decision Tree': Pipeline([
        ('model', DecisionTreeRegressor(random_state=42))
    ]),
    
    'Random Forest': Pipeline([
        ('model', RandomForestRegressor(n_estimators=100, random_state=42))
    ]),
    
    'XGBoost': Pipeline([
        ('model', XGBRegressor(
            n_estimators=200,
            learning_rate=0.1,
            max_depth=6,
            subsample=0.8,
            colsample_bytree=0.8,
            random_state=42,
            n_jobs=-1
        ))
    ])
}

#4. Train and Evaluate Each Pipeline
import pandas as pd

results = []

for name, pipe in pipelines.items():
    pipe.fit(x_train, y_train)
    y_pred = pipe.predict(x_test)
    
    results.append({
        'Model': name,
        'MSE': mean_squared_error(y_test, y_pred),
        'MAE': mean_absolute_error(y_test, y_pred),
        'R2': r2_score(y_test, y_pred)
    })

results_df = pd.DataFrame(results)
print(results_df)
```

```python
# === Import libraries ===
import pandas as pd
import joblib
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# Models
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor

# === Step 1: Train-Test Split ===
x_train, x_test, y_train, y_test = train_test_split(
    df_scaled_x, y, test_size=0.2, random_state=42
)

# === Step 2: Define Model Pipelines ===
pipelines = {
    'Linear Regression': Pipeline([
        ('model', LinearRegression())
    ]),
    
    'Decision Tree': Pipeline([
        ('model', DecisionTreeRegressor(random_state=42))
    ]),
    
    'Random Forest': Pipeline([
        ('model', RandomForestRegressor(
            n_estimators=100,
            random_state=42,
            n_jobs=-1
        ))
    ]),
    
    'XGBoost': Pipeline([
        ('model', XGBRegressor(
            n_estimators=500,
            learning_rate=0.05,
            max_depth=6,
            subsample=0.9,
            colsample_bytree=0.9,
            random_state=42,
            n_jobs=-1
        ))
    ])
}

# === Step 3: Train, Evaluate, and Store Metrics ===
results = []

for name, pipe in pipelines.items():
    pipe.fit(x_train, y_train)
    y_pred = pipe.predict(x_test)
    
    mse = mean_squared_error(y_test, y_pred)
    mae = mean_absolute_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    
    results.append({
        'Model': name,
        'MSE': mse,
        'MAE': mae,
        'R2': r2
    })

results_df = pd.DataFrame(results)
print("\n📊 Model Performance Comparison:")
print(results_df)

# === Step 4: Identify the Best Model (highest R² or lowest MSE) ===
best_model_row = results_df.loc[results_df['R2'].idxmax()]
best_model_name = best_model_row['Model']
best_pipeline = pipelines[best_model_name]

print(f"\n🏆 Best Model: {best_model_name}")
print(f"MSE: {best_model_row['MSE']:.6f}, MAE: {best_model_row['MAE']:.6f}, R2: {best_model_row['R2']:.6f}")

# === Step 5: Save the Best Model as .pkl ===
joblib.dump(best_pipeline, f'best_model_{best_model_name.replace(" ", "_")}.pkl')
print(f"\n✅ Model saved successfully as 'best_model_{best_model_name.replace(' ', '_')}.pkl'")

# === Optional Step 6: Load and Test Later ===
# loaded_model = joblib.load('best_model_Random_Forest.pkl')
# y_pred_loaded = loaded_model.predict(x_test)
```
